```python
%%sql
UPDATE gold.fact_order
SET type_ =
    CASE
        -- ✅ Đơn chưa bị hủy
        WHEN status_key <> 2 THEN 'NOT CANCEL'

        -- ✅ Auto cancel do user không trả tiền
        WHEN LOWER(cancel_comment) LIKE '%auto cancel, user not pay%'
        THEN 'Auto cancel, user not pay'

        -- ✅ Auto cancel - hệ thống tự hủy, không có tài xế
        WHEN LOWER(cancel_comment) LIKE '%auto cancel%'
             OR LOWER(cancel_comment) LIKE '%hệ thống tự động hủy%'
        THEN 'Auto cancel, no driver accepted'

        -- ✅ Không có tài xế nhận đơn
        WHEN LOWER(cancel_comment) LIKE '%no driver%'
             OR LOWER(cancel_comment) LIKE '%driver no accept%'
             OR LOWER(cancel_comment) LIKE '%k có ship%'
             OR LOWER(cancel_comment) LIKE '%k có tài xế%'
             OR LOWER(cancel_comment) LIKE '%k cos ship%'
             OR LOWER(cancel_comment) LIKE '%k liên hệ đk vs tài xế%'
             OR LOWER(cancel_comment) LIKE '%kco taif xế%'
             OR LOWER(cancel_comment) LIKE '%khong có tai xe%'
             OR LOWER(cancel_comment) LIKE '%ko có tx%'
             OR LOWER(cancel_comment) LIKE '%không có xe%'
             OR LOWER(cancel_comment) LIKE '%ko co shipper%'
             OR LOWER(cancel_comment) LIKE '%ko có tài xế%'
             OR LOWER(cancel_comment) LIKE '%ko có tài xế nhận%'
             OR LOWER(cancel_comment) LIKE '%ko có tài xế nhận đơn%'
             OR LOWER(cancel_comment) LIKE '%ko thay xe%'
             OR LOWER(cancel_comment) LIKE '%tài xế ko tới%'
             OR LOWER(cancel_comment) LIKE '%không có tài xế nhận đơn%'
        THEN 'Driver no accept'

        -- ✅ Đặt lại thời gian giao hàng
        WHEN LOWER(cancel_comment) LIKE '%user changes order time%'
             OR LOWER(cancel_comment) LIKE '%hẹn lại%'
             OR LOWER(cancel_comment) LIKE '%reschedule%'
             OR LOWER(cancel_comment) LIKE '%reschedules%'
             OR LOWER(cancel_comment) LIKE '%mai giao%'
             OR LOWER(cancel_comment) LIKE '%đổi thời gian%'
             OR LOWER(cancel_comment) LIKE '%đổi giờ%'
             OR LOWER(cancel_comment) LIKE '%doi thoi gian%'
             OR LOWER(cancel_comment) LIKE '%doi gio%'
        THEN 'Reschedule'

        -- ✅ Người dùng thay đổi thông tin hoặc nhập sai
        WHEN LOWER(cancel_comment) LIKE '%change%'
             OR LOWER(cancel_comment) LIKE '%wrong%'
             OR LOWER(cancel_comment) LIKE '%in correct%'
             OR LOWER(cancel_comment) LIKE '%doesn%'
             OR LOWER(cancel_comment) LIKE '%thay đổi%'
             OR LOWER(cancel_comment) LIKE '%đổi ý%'
             OR LOWER(cancel_comment) LIKE '%sai%'
             OR LOWER(cancel_comment) LIKE '%chọn lại%'
             OR LOWER(cancel_comment) LIKE '%tạo đơn hàng sai%'
        THEN 'User changes something/Wrong information'

        -- ✅ Người dùng chưa có hàng
        WHEN LOWER(cancel_comment) LIKE '%user has no item%'
             OR LOWER(cancel_comment) LIKE '%unavailable package%'
             OR LOWER(cancel_comment) LIKE '%no available items%'
             OR LOWER(cancel_comment) LIKE '%tôi chưa có hàng để giao%'
        THEN 'User has no item'

        -- ✅ Người dùng đổi app khác
        WHEN LOWER(cancel_comment) LIKE '%user used another app%'
             OR LOWER(cancel_comment) LIKE '%đổi nhà vận chuyển%'
             OR LOWER(cancel_comment) LIKE '%grab%'
        THEN 'User used another app'

        -- ✅ Tài xế yêu cầu hủy
        WHEN LOWER(cancel_comment) LIKE '%driver ask%'
             OR LOWER(cancel_comment) LIKE '%lái xe yêu cầu%'
             OR LOWER(cancel_comment) LIKE '%tài xế yêu cầu%'
        THEN 'Driver asked to cancel'

        -- ✅ Hủy bởi đối tác / POS / partner
        WHEN LOWER(cancel_comment) LIKE '%cancel by partner%'
             OR LOWER(cancel_comment) LIKE '%yêu cầu hủy đơn hàng từ hệ thống nhanh.vn%'
             OR LOWER(cancel_comment) LIKE '%ecom%'
             OR LOWER(cancel_comment) LIKE '%boxme%'
             OR LOWER(cancel_comment) LIKE '%khách hàng hủy%'
             OR LOWER(cancel_comment) LIKE '%pos_shop%'
             OR LOWER(cancel_comment) LIKE '%cancel by syste%'
             OR LOWER(cancel_comment) LIKE '%yêu cầu hủy%'
        THEN 'User cancel (POS, partner)'

        -- ✅ Thời gian chờ lâu do gói hàng
        WHEN LOWER(cancel_comment) LIKE '%long package waiting time%'
        THEN 'Long package waiting time'

        -- ✅ Lý do liên quan khuyến mãi
        WHEN LOWER(cancel_comment) LIKE '%promotion%'
             OR LOWER(cancel_comment) LIKE '%promo%'
        THEN 'Related promotion/promo code'

        -- ✅ Vấn đề liên quan COD
        WHEN LOWER(cancel_comment) LIKE '%cod is not affordable%'
             OR LOWER(cancel_comment) LIKE '%cod%'
        THEN 'COD related problems'

        -- ✅ Xe không phù hợp
        WHEN LOWER(cancel_comment) LIKE '%vehicle is unsuitable%'
             OR LOWER(cancel_comment) LIKE '%baga%'
        THEN 'Vehicle is unsuitable'

        -- ✅ Tài xế đến trễ
        WHEN LOWER(cancel_comment) LIKE '%driver took too long to arrive%'
             OR LOWER(cancel_comment) LIKE '%driver pick order too many time%'
             OR LOWER(cancel_comment) LIKE '%long waiting time for driver to arrive%'
        THEN 'Driver took too long to arrive'

        -- ✅ Chờ tài xế nhận quá lâu
        WHEN LOWER(cancel_comment) LIKE '%long waiting time for driver to accept%'
             OR LOWER(cancel_comment) LIKE '%driver took too long to accept the order%'
             OR LOWER(cancel_comment) LIKE '%thời gian chờ lái xe chấp nhận quá lâu%'
        THEN 'Long waiting time for driver to accept'

        -- ✅ Gói hàng quá to
        WHEN LOWER(cancel_comment) LIKE '%package is too large%'
             OR LOWER(cancel_comment) LIKE '%too bulky%'
        THEN 'Package is too large/Packages are too bulky to carry'

        -- ✅ Tài xế ở quá xa
        WHEN LOWER(cancel_comment) LIKE '%driver is so far%'
             OR LOWER(cancel_comment) LIKE '%too far from driver%'
             OR LOWER(cancel_comment) LIKE '%driver is far away%'
        THEN 'Driver was too far from pickup point'

        -- ✅ Tài xế chưa sẵn sàng
        WHEN LOWER(cancel_comment) LIKE '%driver''s not ready%'
             OR LOWER(cancel_comment) LIKE '%driver is not ready%'
        THEN 'Driver''s not ready'

        -- ✅ Các loại Cancel by... khác
        WHEN LOWER(cancel_comment) LIKE '%cancel by user%'
             OR cancel_by_user = TRUE
        THEN 'Cancel by user'

        WHEN LOWER(cancel_comment) LIKE '%cancel by supplier%'
        THEN 'Cancel by supplier'

        WHEN LOWER(cancel_comment) LIKE '%cancel by admin for ahamove%'
        THEN 'Cancel by admin for ahamove'

        WHEN LOWER(cancel_comment) LIKE '%cancel by admin for supplier%'
        THEN 'Cancel by admin for supplier'

        WHEN LOWER(cancel_comment) LIKE '%cancel by admin for user%'
        THEN 'Cancel by admin for user'

        WHEN LOWER(cancel_comment) LIKE '%cancel by merchant bot for partner%'
             OR LOWER(cancel_comment) LIKE '%nguoi%'
        THEN 'Cancel by merchant bot for partner'

        -- ✅ Mặc định
        ELSE 'Others'
    END;
```

```python
%%sql
UPDATE gold.fact_order
SET person_cancel = 
    CASE 
        -- Chỉ xét các đơn chưa hoàn tất
        WHEN complete_date IS NULL THEN
            CASE
                -- ✅ By system
                WHEN LOWER(cancel_comment) LIKE '%auto cancel%'
                     OR LOWER(cancel_comment) LIKE '%long package%'
                     OR LOWER(cancel_code) LIKE '%ahamove%'
                     OR LOWER(cancel_comment) LIKE '%cancel by admin%'
                     OR LOWER(type_) LIKE '%cancel by admin%'
                     OR LOWER(cancel_comment) LIKE '%hệ thống tự động hủy%' 
                THEN 'By system'

                -- ✅ By user
                WHEN cancel_by_user = TRUE
                     OR LOWER(cancel_comment) LIKE '%cannot be reached%'
                     OR LOWER(cancel_comment) LIKE '%cancel booking%'
                     OR LOWER(cancel_comment) LIKE '%drop-off location%'
                     OR LOWER(cancel_comment) LIKE '%too bulky%'
                     OR LOWER(cancel_comment) LIKE '%no show%'
                     OR LOWER(cancel_comment) LIKE '%is not affordable%'
                     OR LOWER(cancel_comment) LIKE '%pick-up location doesn%'
                     OR LOWER(cancel_comment) LIKE '%reschedules the delivery to a different time%'
                     OR LOWER(cancel_comment) LIKE '%k có ship%'
                     OR LOWER(cancel_comment) LIKE '%k có tài xế%'
                     OR LOWER(cancel_comment) LIKE '%k cos ship%'
                     OR LOWER(cancel_comment) LIKE '%k liên hệ đk vs tài xế%'
                     OR LOWER(cancel_comment) LIKE '%kco taif xế%'
                     OR LOWER(cancel_comment) LIKE '%khong có tai xe%'
                     OR LOWER(cancel_comment) LIKE '%ko có tx%'
                     OR LOWER(cancel_comment) LIKE '%không có xe%'
                     OR LOWER(cancel_comment) LIKE '%ko co shipper%'
                     OR LOWER(cancel_comment) LIKE '%ko có tài xế%'
                     OR LOWER(cancel_comment) LIKE '%ko có tài xế nhận%'
                     OR LOWER(cancel_comment) LIKE '%ko có tài xế nhận đơn%'
                     OR LOWER(cancel_comment) LIKE '%ko thay xe%'
                     OR LOWER(cancel_comment) LIKE '%tài xế ko tới%'
                     OR LOWER(cancel_comment) LIKE '%huy%'
                     OR LOWER(type_) LIKE '%driver no accept%'
                     OR LOWER(type_) LIKE '%cancel by user%'
                     OR LOWER(type_) LIKE '%driver took too long to arrive%'
                     OR LOWER(cancel_comment) LIKE '%user%'
                THEN 'By user'

                -- ✅ By supplier
                WHEN LOWER(cancel_comment) LIKE '%for supplier%'
                     OR LOWER(cancel_comment) LIKE '%driver ask%'
                     OR LOWER(cancel_comment) LIKE '%reject to pick up%'
                     OR LOWER(cancel_comment) LIKE '%by supplie%'
                     OR LOWER(cancel_comment) LIKE '%by driver%'
                     OR LOWER(cancel_comment) LIKE '%by supplier%'
                     OR LOWER(cancel_comment) LIKE '%pick up at location%'
                     OR LOWER(cancel_comment) LIKE '%vehicle is%'
                     OR LOWER(cancel_comment) LIKE '%not ready%'
                     OR LOWER(cancel_comment) LIKE '%ship bao huy%'
                     OR LOWER(cancel_comment) LIKE '%tài xế gọi hủy đơn%'
                     OR LOWER(cancel_comment) LIKE '%tai xe hong xe%'
                     OR LOWER(cancel_comment) LIKE '%location is too far from driver%'
                THEN 'By supplier'

                -- ✅ By others (mặc định)
                ELSE 'By others'
            END

        -- Nếu đơn đã complete thì để NULL
        ELSE NULL
    END;
```

```python
%%sql
UPDATE gold.fact_order
    SET customer_group =
    CASE
        -- ✅ GHN Group
        WHEN LOWER(partner) IN ('ghn', 'ghnlastmile') THEN 'GHN'

        -- ✅ Warehouse (WH)
        WHEN LOWER(partner) LIKE '%ka_warehouse%' THEN 'WH'

        -- ✅ Marketplace (MP)
        WHEN LOWER(partner) IN (
            'ghnbulky', 'ghnfreight', 'ghnsameday',
            'shopee', 'spe_reverse', 'spefresh',
            'tiki', 'lazada', 'vugiafarm',
            'ghnbulkytts', 'ghnbulkylastmile', 'ghnbulkysme'
        ) THEN 'MP'

        -- ✅ SMEs (empty or blank partner)
        WHEN TRIM(partner) = '' OR partner IS NULL THEN 'SMEs'

        -- ✅ Key Account (all others except ahamove)
        WHEN LOWER(partner) <> 'ahamove' THEN 'Key Account'

        -- ✅ Default
        ELSE 'SMEs'
    END;
--FROM gold.fact_order;
```

```python
%%sql

CREATE OR REPLACE TABLE gold.Order_route AS
SELECT
    fo.order_id,

    -- ✅ From City: district_key của bản ghi không có stopstatus_key
    MAX(CASE WHEN fo.stopstatus_key IS NULL THEN fo.district_key END) AS from_city,

    -- ✅ To City: district_key của bản ghi có stopstatus_key
    MAX(CASE WHEN fo.stopstatus_key IS NOT NULL THEN fo.district_key END) AS to_city,
    
    -- Distance
    MAX(fo.distance) AS distance,

    -- ✅ Order Status (lấy status_key)
    MAX(fo.status_key) AS order_status,

    -- ✅ JOIN DimDistrict để lấy thông tin "From District"
    MAX(CASE WHEN fo.stopstatus_key IS NULL THEN dd_from.district END) AS from_district_name,
    MAX(CASE WHEN fo.stopstatus_key IS NULL THEN dd_from.latitude END) AS from_lat,
    MAX(CASE WHEN fo.stopstatus_key IS NULL THEN dd_from.longitude END) AS from_lon,

    -- ✅ JOIN DimDistrict để lấy thông tin "To District"
    MAX(CASE WHEN fo.stopstatus_key IS NOT NULL THEN dd_to.district END) AS to_district_name,
    MAX(CASE WHEN fo.stopstatus_key IS NOT NULL THEN dd_to.latitude END) AS to_lat,
    MAX(CASE WHEN fo.stopstatus_key IS NOT NULL THEN dd_to.longitude END) AS to_lon

FROM gold.fact_order fo
-- ✅ JOIN cùng 2 lần với DimDistrict (1 cho from, 1 cho to)
LEFT JOIN gold.dim_district dd_from 
    ON fo.district_key = dd_from.district_key
LEFT JOIN gold.dim_district dd_to 
    ON fo.district_key = dd_to.district_key

GROUP BY fo.order_id;
```

```python
%%sql

CREATE OR REPLACE TABLE gold.dim_supplier AS
SELECT *,
    CASE
        WHEN age <= 22 THEN '<= 22'
        WHEN age >= 23 AND age <= 35 THEN '23-35'
        WHEN age >=36  AND age <= 50 THEN '36-50'
        WHEN age > 50 THEN '>50'
        ELSE 'Unknown'
    END AS age_group
FROM gold.dim_supplier;
```
